```python
import pandas as pd
```

```python
#читаем файл
df = pd.read_csv("Customers.csv", sep=";")

#тестовый вывод таблицы
df
```

```python
#находим пропуски при помощи .isna(), а также их количество при помощи .sum()
df.isna().sum()
```

### 1 вариант (удаление)

```python
#так как пропусков не так много, то можно просто удалить строки с пропусками
df_fixed_drop = df.dropna()
df_fixed_drop
```

### 2 вариант (заполнение)

```python
#конечно, в данном случае, такое заполнение данных не принесет особой пользы,
#кроме обхода ошибок, поэтому лучше было бы заполнить эти данные индивидулально
#Да, можно было бы написать "Без работы", но мы не можем гарантировать правильность этих данных
df_fixed_fill = df.fillna("Неизвестно")
df_fixed_fill[df_fixed_fill["Profession"] == "Неизвестно"]
```

```python
#groupby("Profession") групирует данные по профессии

#["Income"] указывает, что мы хотим работать с колонкой "Income"

#mean находит средние значения
df_fixed_fill.groupby('Profession')["Income"].mean()
```
