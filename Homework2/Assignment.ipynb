```python
import matplotlib.pyplot as plt
```

```python
# Наблюдения
X = [1, 3, 7]
y = [2, 6, 14]

# скорость обучения
lr = 0.01

# начальные веса
w1 = 0
w0 = 0

for i in range(len(X)):
    prediction = w1 * X[i] + w0
    w1 = w1 + 2 * lr * X[i] * (y[i] - prediction)
    w0 = w0 + 2 * lr * (y[i] - prediction)

print('w0 =', w0)
print('w1 =', w1)
```

```python
#наблюдения
plt.scatter(X, y)

#результаты обучения
plt.plot(X, [w1 * x + w0 for x in X], color="red")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.title("График обучения")
```
