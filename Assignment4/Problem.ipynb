2. Найдите людей, у которых возраст больше 30 и доход меньше 30000, 

найдите людей, которые по профессии юристы и с опытом работы больше 5 лет
<hr>

```python
import pandas as pd
```

```python
#читаем файл
df = pd.read_csv("Customers.csv", sep=";")

#тестовый вывод таблицы
df
```

```python
#находим людей, старше 30, чей доход меньше 30 000 рублей
df[(df["Age"] > 30) & (df['Income'] < 30000)]
```

```python
#люди, которые по профессии юристы и с опытом работы больше 5 лет
df[(df["Profession"] == "Lawyer") & (df["Work Experience"] > 5)]
```
