1. Сгенерируйте искусственные данные(один признак и одна целевая переменная) и разделите данные на обучающую и тестовую выборки(функция train_test_split)

<hr>

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(20)
X = 2 * np.random.rand(100, 1)  # один признак
y = 4 + 3 * X + np.random.randn(100, 1)  # целевая переменная с небольшим шумом

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

2. Создайте и обучите модель линейной регрессии, после чего сделайте предсказания 
<hr>

```python
model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

3. Постройте точечные графики исходных данных(выделите тестовые и обучающие данные разным цветом) и график линии регрессии, где по оси абсцисс - признак, по оси ординат - целевая переменная. Все это должно быть на одних осях 
<hr>

```python
plt.figure(figsize=(10,5))

# график исходных данных

plt.scatter(X_train, y_train, color='skyblue', label="Обущающий набор")
plt.scatter(X_test, y_test, color='purple', label="Тестовый набор", alpha=0.6)

#график линейной регрессии

plt.plot(X_test, y_pred, color="red", label='Линейная регрессия')
plt.grid()
plt.legend()
plt.xlabel("Признак")
plt.ylabel("Целевая переменная")
plt.title("График модели")
plt.show()
```
