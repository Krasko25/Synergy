Количество эпох имеет смысл брать не меньше 100, а lr подбирайте самостоятельно. lr считается подобранным верно, если получившиеся веса отличаются от идеальных значений w1 = 3, w2 = 2, b  = -1 не более, чем на 1. 
Подсказка: Если веса получаются слишком большими по модулю или выводится сообщение об ошибке, стоит уменьшит lr. Если все веса получаются сильно меньше 1, стоит увеличить lr.

```python
#набор данных
x1 = np.random.randint(100, size = 100)
x2 = np.random.randint(200, size = 100)
y = 3 * x1 + 2 * x2 - 1
x1 = x1 + np.random.randint(5, size = 100) / 10
x2 = x2 + np.random.randint(5, size = 100) / 10
y = y + np.random.randint(5, size = 100) / 10
```

```python
#задаем начальные значение
w1 = 0
w2 = 0
w0 = 0

# learning rate был выбран именно такой, так как из-за сложных входных данных 
# программа испытывает перегрузку при больших значениях lr. 
# также такое значение позволило получить наиболее близкий результат
lr = 0.00001

# приблизиться к реальным значениям получилось только при помощи такого большого количества эпох
epochs = 3000

for epoch in range(1, epochs + 1):
    for i in range(len(x1)):
        prediction = w1 * x1[i] + w2 * x2[i] + w0
        w1 += 2 * lr * x1[i] * (y[i] - prediction)
        w2 += 2 * lr * x2[i] * (y[i] - prediction)
        w0 += 2 * lr * (y[i] - prediction)
    
print(f'w1 = {w1}, \nw2 = {w2}, \nw0 = {w0}')
```
