**Задача 3**.
Возможные результаты забега школьников на 100 метров можно сгенерировать следующим образом:

data = np.random.normal(16, 2, 1000)

Постройте гистограмму красного цвета с полупрозрачными столбиками. Какие выводы можно сделать о результатах по ней?

<hr>

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#возможные результаты забега школьников
data = np.random.normal(16, 2, 1000)

#подписывание осей
plt.xlabel("Время (с)")
plt.ylabel("Количество человек")

plt.hist(data, bins=20, color='r', alpha=0.5)

plt.show()
```

На основе этой гистограммы можно сделать выводы о общей физической подготовке участников, о средних значениях, количесве физически отстающих и выдающихся школьников.
