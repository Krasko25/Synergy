```python
#создание сетки
plt.grid()

#подписывание осей
plt.xlabel("X")
plt.ylabel("Y")

#подпись графика
plt.title("Задача 2")

#задаем значения для x и y
X = np.random.normal(0, 1, 3000)
Y = np.random.normal(3, 4, 3000)

#случайный цвет и размер
color = np.random.rand(3000)
size = np.random.rand(3000) * 100

#c - цвет, s - размер
plt.scatter(X, Y, c=color, s=size, alpha=0.3, marker="*")
```
